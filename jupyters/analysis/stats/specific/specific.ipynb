### Save current dataframe to csv where are situated all stats runs

```python
from os import path
```

```python
def append_to_file(input_df, filename=NAME_OF_STATS_FILE):
    if path.exists(filename):
        input_df.to_csv(filename, mode='a', index=False, header=False)
    else:
        input_df.to_csv(filename, mode='a', index=False, header=True)
```

```python
append_to_file(current_df)
```
